__Задание 1__. В файле `scores.npy` содержится массив оценок рейтингов вопросов. Напишите функцию, которая считает сумму квадратов цифр числа. Векторизуйте эту функцию при помощи `numba.vectorize` и посчитайте количество чисел из файла, сумма квадратов цифр которых меньше 50.

```python
import numpy as np
import pandas as pd
```

```python
scores = np.load('scores.npy')
scores
```

```python
# функция считает сумму квадратов цифр числа
def sum_sqrt_number(x):
    x = abs(x)
    summ = 0
    while x > 0:
        summ += (x%10)**2
        x = x//10
    return summ
```

```python
sum_sqrt_number(22)
```

```python
sum_sqrt_number(0)
```

```python
sum_sqrt_number(-3)
```

```python
import numba

# Векторизуруем эту функцию при помощи numba.vectorize

vector = numba.vectorize(["int64(int64)"])(sum_sqrt_number)
```

```python
vector(scores)
```

```python
# cчитаем количество чисел из файла, сумма квадратов цифр которых меньше 50.
(vector(scores) < 50).sum()
```

Напишите функцию, которая для вопроса с индексом `i` выводит заголовки 5 ближайших вопросов в смысле расстояния между эмбеддингами этих вопросов. Выберите метрику, наиболее подходящую (на ваш взгляд) для решения этой задачи. Для работы с эмбеддингами используйте `dask.array`.

```python
# считали файл
import h5py

hdf = h5py.File("qembeddinbgs.h5", "r")
dset = hdf["embeddings"]
dset

# Я открыл с помощью проги специальной посмотрел, которая h5 файлы читает
```

```python
# преобразовали файл в dask
import dask.array as da

arr = da.from_array(dset)
arr
```

```python
# считали другой файл
import pickle

with open('qtitles.pickle', 'rb') as f:
    qtitles = pickle.load(f)
```

```python
# чтобы найти расстояние Чебышева нужно из одного вектора вычесть другой и найти максимум из получившихся значений
# из всех векторов вычитаем вектор i и ищем максимум
# получается массив с расстояниями от каждого элемента до i-го
# потом я i-му элементу массива присваиваю максимум 
# (до этого он равен нулю, тк расстояние вектора до самого себя 0) чтобы оно нам не мешалось дальше.
# потом я создаю список при помощи da.argsort и потом ищу в цикле 5 минимальных элементов
```

```python
def cheb(i): #расстояние Чебышева
    cheb_array = da.max(arr - arr[:, i].reshape(-1, 1), axis = 0) # создаеv массив с расстояниями до i элемента по формуле расстояния чебышева
    cheb_array[i] = cheb_array.max() # чтобы не мешалось
    cheb_array = list(np.argsort(cheb_array)) # возвращает индексы, сортирующие элементы исходного массива
    for j in range(5):
        print(qtitles[cheb_array.index(j)])
```

```python
import dask.bag as db
import json

q = db.read_text('D:\\que\\questions_with_answers\\questions_with_answers_*.jsonl')
q.take(5)
```

```python
# пытаюсь обработать обычным способом через json.loads, а если не получается заменяю на словарь
def fix_json_load(bag):
    try:
        return json.loads(bag)
    except json.JSONDecodeError:
        return {'Id': 0, 'answers': ['NAN(']}
```

```python
q = q.map(fix_json_load) # map()чтобы применить функцию к каждому элементу в итерируемом объекте
q.take(5)
```

```python
def three_groups(x):
    if x['answers'] == ['NAN(']:
        return 'NaN'
    else:
        if len(x['answers']) < 5:
            return '<5'
        elif len(x['answers']) >= 5 and len(x['answers']) < 15:
            return '>=5 & <15'
        else:
            return '>=15'
```

```python
from collections import Counter

Counter(list(q.map(three_groups)))

#Класс Counter() модуля collections - это подкласс словаря dict для подсчета хеш-объектов (неизменяемых, таких как строки, числа, кортежи и т.д.). 
#Это коллекция, в которой элементы хранятся в виде словарных ключей, а их счетчики хранятся в виде значений словаря.
```
